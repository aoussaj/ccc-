```python
import numpy as np
import matplotlib.pyplot as plt

# Load the Duschinsky matrix
filename = r"C:\fcclasses3-3.0.4\fcclasses3-3.0.4\absAH300Cartesianzerodisplacement\duschinsky.dat"
data = np.loadtxt(filename)

# Determine the matrix dimension
N = int(np.sqrt(data.size))
if N * N != data.size:
    raise ValueError("Matrix size is incorrect.")

# Reshape into an NxN matrix
matrix = data.reshape((N, N))

# Identify high values in the matrix (highlighted in the red circle)
threshold = 0.5  # Adjust based on intensity in the image
indices = np.argwhere(np.abs(matrix) > threshold)  # Find (GS, ES) mode pairs

# Print the corresponding GS and ES modes
print("Significant Mode Pairs (GS mode, ES mode):")
for gs_mode, es_mode in indices:
    print(f"({gs_mode + 1}, {es_mode + 1})")  # +1 to match mode numbering (1 to 123)

# Plot the matrix
plt.figure(figsize=(8, 8))
plt.imshow(matrix, cmap='viridis', origin='upper', interpolation='none')
plt.colorbar(label='G(I,J)')
plt.xlabel("ES modes")
plt.ylabel("GS modes")
plt.title("Duschinsky Matrix (AH_cartesian)")
plt.xticks([0, N-1], [1, N])
plt.yticks([0, N-1], [1, N])

# Highlight detected mode pairs
for gs_mode, es_mode in indices:
    plt.scatter(es_mode, gs_mode, color='red', s=10)

plt.show()
```
